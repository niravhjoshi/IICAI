```python
import requests
from urllib.request import Request, urlopen
from urllib.request import urlopen, URLError, HTTPError
import urllib.error
import json
```

```python
user_iics_loginURL=' https://dm-us.informaticacloud.com/ma/api/v2/user/login'

headers = {
        'Content-Type': "application/json",
        'Accept': "application/json",
        'cache-control': "no-cache"
        
        }

payload = "{\r\n\"@type\": \"login\",\r\n\"username\": \"niravjoshi\",\r\n\"password\": \"Informatica123\"\r\n}"

response = requests.request("POST", user_iics_loginURL, data=payload, headers=headers)

#print(response.text)
resp_obj = json.loads(response.text)
session_id = resp_obj['icSessionId']
print(session_id)

#Now using this session we are calling activity log 
activity_url = "https://na1.dm-us.informaticacloud.com/saas/api/v2/activity/activityLog"
act_headers = {
    'icSessionId': session_id,
    'Content-Type': "application/json",
    'cache-control': "no-cache",
    
    }
print(act_headers)

act_response = requests.request("GET", activity_url,  headers=act_headers)
#print(act_response.text)
print("Now loading Activity Log Data:-")
act_json_data= json.loads(act_response.text)
print(act_json_data)
```
